```python
#Import neccesary libraries
import pandas as pd 
import numpy as np 
```

```python
#Import data using pandas 
df= pd.read_csv('/kaggle/input/used-cars-prices-in-uk/used_cars_UK.csv')
df.info()
```

```python
#Exploring data 
df.head(5)
```

```python
df.describe()
```

```python
#Checking data validation
null_values = df.isnull().sum()
null_values
```

```python
#filling null value for doors with the mean 
door_means = df['Doors'].mean()
door_mean= door_means.round().astype(int)
doors = df['Doors'].fillna(door_mean,inplace=True)
df.isnull().sum()
```

```python
#filling null value for seats with the mean 
seat_means = df['Seats'].mean()
seat_mean= seat_means.round().astype(int)
seat = df['Seats'].fillna(seat_mean,inplace=True)
df.isnull().sum()
```

```python
duplicated_rows = df.duplicated().sum()
duplicated_rows
```

```python
#importing sklearn models 
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
lr=linear_model.LinearRegression()
lr
```

```python
df.columns
```

```python
#Checking correlations 
variables = df[['Price','Mileage(miles)','Registration_Year','Doors', 'Seats',]]
variables.corr()['Price']
```

```python
df_2 = df[['Price','Mileage(miles)','Registration_Year','Doors','Seats']]
df_2.head(10)
```

```python
independent_variables = df[['Mileage(miles)','Registration_Year','Doors','Seats']].to_numpy()
independent_variables
```

```python
target_variable = df['Price'].to_numpy()
target_variable
```

Training The Model 

```python
x=independent_variables 
y=target_variable
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=500, random_state=5)
```

```python
y_train
```

```python
lr.fit(x_train,y_train)
```

```python
correlaation_coefficient = lr.coef_
correlaation_coefficient
```

```python
y_prediction = lr.predict(x_test)
y_prediction
df_3 = pd.DataFrame(y_prediction, columns=['Prediction'])
df_3.head(100)
```

```python
df_2['Price Prediction']=df_3
```
